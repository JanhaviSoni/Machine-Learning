# Polynomial Linear Regression

# Importing Libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing Data set

```python
dataset = pd.read_csv('../Machine_Learning/Position_Salaries.csv')
# here we have features and dependent variable
X= dataset.iloc[:,1:-1].values # we are not taking the column of names of various positions like manager etc as it is easy to 
                               #see them
y= dataset.iloc[:,-1].values 
```

```python
print(X)
```

```python
print(y)
```

**Here we are not splitting into training and test set as dataset is small . We will predict for value 6.5**

# Training of Linear Regression model on whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
y_pred = lin_reg.predict(X)
```

# Training of Polynomial Regression model on whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

# Visualising Linear Regression model

```python
# plotting real value points
plt.scatter(X,y, color = 'red')
#plotting predicted values
plt.plot(X,y_pred, color='blue')
plt.title("Truth or Bluff Linear Regression")
plt.xlabel('Position')
plt.ylabel('Salaries')
plt.show()
```

# Visualising Polynomial Regression model

```python
# plotting real value points
plt.scatter(X,y, color = 'red')
#plotting predicted values
plt.plot(X,lin_reg_2.predict(X_poly), color='blue')
plt.title("Truth or Bluff Polynomial Regression")
plt.xlabel('Position')
plt.ylabel('Salaries')
plt.show()
```

# Training Polynomial Regression for Higher Degree

```python
from sklearn.preprocessing import PolynomialFeatures
X_poly_reg = PolynomialFeatures(degree = 4)
new_X_poly = X_poly_reg.fit_transform(X)
higher_lin_reg = LinearRegression()
higher_lin_reg.fit(new_X_poly,y)
```

# Visualising Polynomial Regression for higher degree

```python
# plotting real value points
plt.scatter(X,y, color = 'red')
#plotting predicted values
plt.plot(X,higher_lin_reg.predict(new_X_poly), color='blue')
plt.title("Truth or Bluff Polynomial Regression")
plt.xlabel('Position')
plt.ylabel('Salaries')
plt.show()
```

# Prediction a new value using Linear Regression

```python
lin_reg.predict([[6.5]]) # an array is represented as [[a,b],[c,d]] 2 rows and 2 columns
```

# Prediction a new value using Polynomial Regression

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
